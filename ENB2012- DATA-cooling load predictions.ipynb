# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Import Data Set and Clean the dataset

```python
data=pd.read_csv('ENB2012_Data.csv')
data.dropna(how='all') #remove NaN values from the data set
datanew= data.dropna(axis=1, how='all') #first deletes all the non useful columns from the dataset
data_processed=datanew.dropna(how='all') #then deletes all the unuseful rows from the data set
print(data_processed)
#lets try re-naming all the headers columns
data_finalized=data_processed.rename(columns={"X1": "Relative Compactness", "X2": "Surface Area","X3": "Wall Area","X4": "Roof Area","X5": "Overall Height","X6": "Orientation","X7": "Glazing Area","X8": "Glazing Area Distribution","Y1": "Heating Load","Y2":"Cooling Load"})
data_finalized.head(10) #lets me look at random samples within the dataset
```

# Define x and y variables

```python
x=data_finalized.iloc[:,:-2].values
y=data_finalized.iloc[:,9:10].values
print(y)
```

```python
#y = y.reshape(len(y),1) #do this only for feature scaling for Support Vector Regressor mode
```

# Split data into training and test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```
